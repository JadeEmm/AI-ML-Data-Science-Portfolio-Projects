# Preparation

```python
df.drop(['Unnamed: 32'], axis =1, inplace=True)
```

```python
df.shape
```

```python
#rename columns

df.rename(columns = {'concave points_se':'concave_points_se',
                     'concave points_worst':'concave_points_se'}, inplace=True)
```

```python
df.columns
```

```python
#Get dummy variable column for target variable
#Create dummies and store in a variable
dummies = pd.get_dummies(df.diagnosis)
#Concatenate
df = pd.concat([df,dummies], axis ='columns')
#Drop 'diagnosis' and 'B' dummy column as not needed (duplicate of 'M')
df.drop(['diagnosis','B'], axis=1, inplace=True)
#rename 'M' dummy column as 'diagnosis: 0 = benign and 1 = malignant
df.rename(columns = {'M':'diagnosis'}, inplace=True)
```

```python
df.head()
```

```python
#Move diagnosis column to the front
cols_to_move =['id','diagnosis']
df = df[cols_to_move + [ col for col in df.columns if col not in cols_to_move] ]
```

```python
#drop id column
df.drop(['id'], axis=1, inplace=True)
```

```python
df.head()
```

# Visualisation

# Support Vector Machine

```python
#Import train_test_split function
from sklearn.model_selection import train_test_split

#Define feature columns and target

X=df.drop(['diagnosis'], axis=1) #feature variables
y=df['diagnosis'] # target variable

#Split dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state= 88)
```

```python
#Import SVM model
from sklearn import svm
#Create svm classifier
clf = svm.SVC(kernel='rbf') #radial basis function kernel
#clf = svm.SVC(kernel='linear') #linear kernel

#Train the model using training data
clf.fit(X_train,y_train)

#Predict on the test dataset
y_pred = clf.predict(X_test)
```

Evaluate SVM model

```python
#Import metrics module
from sklearn import metrics

print('Accuracy:', metrics.accuracy_score(y_test,y_pred)) #How often is the classifier correct?
print('Precision:', metrics.precision_score(y_test,y_pred)) #What % of positive predictions are actually positive?
print('Recall:', metrics.recall_score(y_test,y_pred)) # Of all real positive cases, how many are predicted positive?
```

Hyperparameter Tuning using GridSearch CV

```python
from sklearn.model_selection import GridSearchCV

#define parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma':[1, 0.1, 0.01, 0.001, 0.001],
              'kernel':['rbf','linear']}
grid = GridSearchCV(clf, param_grid, refit=True, verbose = 3)

#fitting the model for grid search
grid.fit(X_train, y_train)
```

```python
#print best parameter after tuning
print(grid.best_params_)

#print how our model looks after hyper-parameter tuning
print(grid.best_estimator_)
```

```python
#Re-run predictions using best parameters
grid_predictions = grid.predict(X_test)

#Print evaluation metrics
print('Accuracy:', metrics.accuracy_score(y_test,grid_predictions)) #How often is the classifier correct?
print('Precision:', metrics.precision_score(y_test,grid_predictions)) #What % of positive predictions are actually positive?
print('Recall:', metrics.recall_score(y_test,grid_predictions)) # Of all real positive cases, how many are predicted positive?
```

# Naive Bayes

```python
#NB assumes that the effect of a particular feature in a class is independent of other features
```

```python
#Import train_test_split function
from sklearn.model_selection import train_test_split

#Define feature columns and target

X=df.drop(['diagnosis'], axis=1) #feature variables
y=df['diagnosis'] # target variable

#Split dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state= 88)
```

```python
#Build a Gausian Naive Bayes
from sklearn.naive_bayes import GaussianNB
#Build a Gaussian Classifier
model = GaussianNB()

#Train model
model.fit(X_train, y_train)
#Predict Output
y_pred = model.predict(X_test)
```

```python
#Import metrics module
from sklearn import metrics

print('Accuracy:', metrics.accuracy_score(y_test,y_pred)) #How often is the classifier correct?
print('Precision:', metrics.precision_score(y_test,y_pred)) #What % of positive predictions are actually positive?
print('Recall:', metrics.recall_score(y_test,y_pred)) # Of all real positive cases, how many are predicted positive?
print('F1:', metrics.f1_score(y_test,y_pred, average= 'weighted'))
```

```python
from sklearn.model_selection import GridSearchCV

#define parameter range
param_grid = {'var_smoothing':np.logspace(0,-9, num=100) ,
              }
grid = GridSearchCV(model, param_grid, refit=True, verbose = 0)

#fitting the model for grid search
grid.fit(X_train, y_train)
```

```python
#print best parameter after tuning
print(grid.best_params_)

#print how our model looks after hyper-parameter tuning
print(grid.best_estimator_)
```

```python
y_pred = grid.predict(X_test)
print('Accuracy:', metrics.accuracy_score(y_test,y_pred)) #How often is the classifier correct?
print('Precision:', metrics.precision_score(y_test,y_pred)) #What % of positive predictions are actually positive?
print('Recall:', metrics.recall_score(y_test,y_pred)) # Of all real positive cases, how many are predicted positive?
print('F1:', metrics.f1_score(y_test,y_pred, average= 'weighted'))
```
